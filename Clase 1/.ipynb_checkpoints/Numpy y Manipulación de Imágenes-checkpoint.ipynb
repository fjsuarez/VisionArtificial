## Indexación

```python
mat = np.arange(0,100).reshape(10,10)
```

```python
mat
```

```python
fil = 4
col = 6
```

```python
mat[fil,col]
```

```python
# Con Slices
mat[:,col]
```

```python
mat[fil,:]
```

# Imágenes y Numpy

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('../Datos/cachorro.jpg')
```

```python
pic
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
pic_red[:, :, 1] = 0    # Cancelar la contribución del verde
pic_red[:, :, 2] = 0    # Cancelar la contribución del azul
```

```python
plt.imshow(pic_red)
```
