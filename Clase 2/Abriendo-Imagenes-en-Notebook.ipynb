```python
img_bgr = cv2.imread('../Datos/cachorro.jpg')
plt.imshow(img_bgr)
```

La imagne ha sido correctamente cargada por `openCV` como un array de `numpy` pero el color de cada pixel está ordenado en BGR.
`matplotlib` espera una imagen en RGB, por lo que para mostrar correctamente la imagen, es necesario cambiar esos canales.
Esta operación puede hacerse usando las funciones de conversión `cv2.cvtColor()` de `openCV` o trabajando directamente con el array de `numpy`.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gris = cv2.imread('../Datos/cachorro.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gris)
```

```python
img_gris = cv2.imread('../Datos/cachorro.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gris,cmap='gray')
```

## Redimensionar Imágenes

```python
img_rgb.shape
# alto, ancho, canales de color
```

```python
                        # ancho, alto
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### Por proporociones (ratios)

```python
w_ratio = 0.5 # ancho
h_ratio = 0.5 # alto
```

```python
nueva_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(nueva_img)
```

```python
%matplotlib qt
plt.imshow(nueva_img)
```

### Invirtiendo Imágenes

```python
%matplotlib inline
```

```python
# Con respecto al eje central en X
nueva_img = cv2.flip(nueva_img,0)
plt.imshow(nueva_img)
```

```python
# Con respecto al eje central en Y
nueva_img = cv2.flip(nueva_img,1)
plt.imshow(nueva_img)
```

```python
# Con respecto a ambos
nueva_img = cv2.flip(nueva_img,-1)
plt.imshow(nueva_img)
```
